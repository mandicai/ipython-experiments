```python
# Get Mandi's favorite playlist tracks
# url_pt = 'https://api.spotify.com/v1/users/123684296/playlists/0ZLq7Ne2YxNkV6nZbS7bDX/tracks' # Mandi 
# url_pt = 'https://api.spotify.com/v1/users/jnj16180340/playlists/4fd8tKLePFPPt9zD85QXg0/tracks' # Nate J
url_pt = 'https://api.spotify.com/v1/users/juliafooote/playlists/0nq2WQILeeoomnftrxdzXW/tracks' # Julia

headers = {"Authorization":"Bearer {}".format(access_token)} # .format is like concatenating, USE THIS
tracks_playlist_mandi = requests.get(url_pt, headers = headers) # Headers define the information that is needed to fetch information r = r.json() 
tracks_playlist_mandi = tracks_playlist_mandi.json() # Converts response to workable JSON file, class = 'dict'

# Store all ID's into the list
def get_id_url(a):
    
    # Empty list to keep track IDs
    id_tracks_playlist = []
    
    for key in a['items']:
        id_tracks_playlist.append(key['track']['id'])
    id_tracks_playlist = ','.join(id_tracks_playlist)

    url_af = 'https://api.spotify.com/v1/audio-features/?ids={}'.format(id_tracks_playlist)
    return url_af

# Store URL with all ID's into a variable
url_af = get_id_url(tracks_playlist_mandi)

# Fetch audio feature data
af_tracks_playlist_mandi = (requests.get(url_af, headers = headers)).json()
print(json.dumps(af_tracks_playlist_mandi, indent=2))
```

```python
# Grab danceability and valence values and store them into a list
valence = [] # Positive mood of a song
danceability = [] # Ability to dance to a song
for key in af_tracks_playlist_mandi['audio_features']:
    danceability.append(key['danceability'])
    valence.append(key['valence'])
    
print(danceability)
print(valence)
```
