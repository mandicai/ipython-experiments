```python
import numpy as np
import pandas as pd
import csv

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bkcharts import Histogram
```

```python
output_notebook()
```

```python
# Bottom row excluding the first 4 columns
with open("Histograms.csv", 'r') as f:
    rows = list(csv.reader(f))

# Wellplate AAAO91F
bottomrow_AAA901F_int = [int(well) for well in rows[0]]
bottomrow_AAA901F_int = pd.DataFrame({'Absorbancy': bottomrow_AAA901F_int})

# Average of all rows across all plates within a batch
allrows_average_int = [float(well) for well in rows[1]]
allrows_average_int = pd.DataFrame({'Absorbancy': allrows_average_int})

histogram_AAA901F_bottomrow = Histogram(bottomrow_AAA901F_int, bins=10, title="AAA901F Bottom Row")
histogram_average = Histogram(allrows_average_int, bins=10, title="Average Bottom Row")
```

```python
show(histogram_AAA901F_bottomrow)
show(histogram_average)
```
