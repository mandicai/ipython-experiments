```python
# Do some binning
def make_plot(title, hist, edges, color, x_axis):
    p = figure(title=title, tools='', plot_width=500, plot_height=500, background_fill_color='#ffffff')
    p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color=color, line_color='white', alpha=0.5)
    
    p.y_range.start = 0
    p.xaxis.axis_label = x_axis
    p.yaxis.axis_label = 'Number of Events'
    p.grid.grid_line_color='white'
    return p

hist, edges = numpy.histogram(data['FSC-A'], bins=50)

s1 = make_plot('Distribution of Events Across FSC-A', hist, edges, '#FF9E16', 'FSC-A')

hist, edges = numpy.histogram(data['SSC-A'], bins=50)

s2 = make_plot('Distribution of Events Across FSC-H', hist, edges, '#009DFC', 'FSC-H')

hist, edges = numpy.histogram(data['FL1-A'], bins=50)

s3 = make_plot('Distribution of Events Across FSC-W', hist, edges, '#9AC43D', 'FSC-W')
```

```python
file = 'Accuri - C6 - A02 Spherotech 8 Peak Beads.fcs'

plots = row(s1, s2, s3)
show(column(Div(text='<h3>' + file + '</h3>'), plots))
```
