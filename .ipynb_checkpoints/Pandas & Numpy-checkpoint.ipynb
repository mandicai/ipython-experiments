```python
import plotly.plotly as py
import plotly.figure_factory as ff
import pandas as pd
import numpy as np
```

```python
meow = np.random.choice([0, 1], size=(384,))
```

```python
index_one_tuple = np.where(meow == 1)
index_zero_tuple = np.where(meow == 0)

print(index_one_tuple)
print(type(index_one_tuple))
print(len(index_one_tuple))

index_one = []
index_zero = []

for thing in index_one_tuple[0]:
    index_one.append(thing)

for thing in index_zero_tuple[0]:
    index_zero.append(thing) 

print(index_one)
print(index_zero)
```

```python
pd_index_one = pd.DataFrame({'Indices': index_one})
pd_index_zero = pd.DataFrame({'Indices': index_zero})
print(pd_index_one)
```

```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.io import push_notebook 
from bkcharts import Histogram
output_notebook()
```

```python
hist = Histogram(pd_index_one, values = "Indices", title = 'Indices of 1', color = '#a764bd', bins = 15)
show(hist)

hist2 = Histogram(pd_index_zero, values = "Indices", title = 'Indices of 1', color = '#f1f3f4', bins = 15)
show(hist2)
```
